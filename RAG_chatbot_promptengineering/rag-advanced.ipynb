# Using cross-encoder for better retrieval because it's a book so we needed better retrieval.

```python
from sentence_transformers import CrossEncoder
cross_encoder = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
```

# Prompt engineering

```python
from sentence_transformers import CrossEncoder

def format_documents(query):
    results = index.query(vector=embeddings.embed_query(query), top_k=25, include_metadata=True)
    pairs = [(query, match['metadata'].get('text', 'No text available')) for match in results['matches']]
    scores = cross_encoder.predict(pairs)
    reranked_results = list(zip(results['matches'], scores))
    reranked_results.sort(key=lambda x: x[1], reverse=True)
    top_10_results = reranked_results[:15]
    formatted_docs = []
    for match, score in top_10_results:
        doc_id = match['id']
        text = match['metadata'].get('text', 'No text available')  # Assuming metadata has a 'text' field
        
        # Create a formatted template
        formatted_doc = f"""
        Relevance Score: {score}
        Content:
        {text}
        """
        
        formatted_docs.append(formatted_doc)
    
    prompt_template = f'''
    You are a helpfull assistant give the answer to the user query given below using the docs suggested below. Make a detailed answer using only the suggested docs and don't hallucinate.
    User={query}
    Suggested_docs={formatted_docs}
    Assistant: '''
    
    # Join all formatted documents into a single input string
    return prompt_template
```

```python
print(format_documents("What is mosfet?"))
```

# Trying to make a chat bot.

## the promt template needs to be fixed.


```python
prompts=[]
responses=[]
```

```python
from sentence_transformers import CrossEncoder
def previous_conversation(prompts,responses):
    conv=''''''
    for i in range(len(responses)):
        formatted_doc=f'''Query:{prompts[i]}   
        Response:{responses[i]}
        '''
        
        conv+=formatted_doc
    return conv
    

def bot_format_documents(query,prompts,responses):
    query+=previous_conversation(prompts,responses)
    results = index.query(vector=embeddings.embed_query(query), top_k=25, include_metadata=True)
    pairs = [(query, match['metadata'].get('text', 'No text available')) for match in results['matches']]
    scores = cross_encoder.predict(pairs)
    reranked_results = list(zip(results['matches'], scores))
    reranked_results.sort(key=lambda x: x[1], reverse=True)
    top_10_results = reranked_results[:15]
    formatted_docs = []
    for match, score in top_10_results:
        doc_id = match['id']
        text = match['metadata'].get('text', 'No text available')  # Assuming metadata has a 'text' field
        
        # Create a formatted template
        formatted_doc = f"""
        Relevance Score: {score}
        Content:
        {text}
        """
        
        formatted_docs.append(formatted_doc)
    previous_conv=previous_conversation(prompts,responses)
    prompt_template = f'''
    You are a helpfull assistant give the answer to the user query given below using the docs suggested below and also the previous conversation made with you which is a part of the user query given as input to you. Make a detailed answer using only the suggested docs and don't hallucinate.
    User={query}
    Suggested_docs={formatted_docs}
    Assistant: '''
    
    # Join all formatted documents into a single input string
    return prompt_template
```

```python
def answer(query):
    formatted_doc=bot_format_documents(query,prompts,responses)
    
    response = model.generate_content(formatted_doc)
    prompts.append(query)
    responses.append(response.text)
    print(response.text)
```

```python
answer("What is a MUX?")
```
